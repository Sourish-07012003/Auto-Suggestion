```python
import numpy as np
import pandas as pd
import textdistance  
import re
from collections import Counter
```

```python
word = []
with open('autocorrect book.txt', 'r', encoding='utf-8') as f:
    data = f.read()
    data = data.lower()
    words = re.findall('\w+', data)
    words += words
```

```python
V = set(words)
```

```python
words_freq_dict = Counter(words)
```

```python
words_freq_dict.most_common(10)
```

```python
Total_words_freq = sum(words_freq_dict.values())


probs = {}
for k in words_freq_dict.keys():
    probs[k] = words_freq_dict[k] / Total_words_freq
```

```python
probs
```

```python
def autocorrect(word):
    word = word.lower()
    if word in probs:
        print('the word is already there', word)
    else:
        similarities = [1-(textdistance.Jaccard(qval=2)).distance(w,word) for w in words_freq_dict.keys()]
        df = pd.DataFrame.from_dict(probs,orient='index').reset_index()
        df = df.rename(columns={'index':'Word',0:'Prob'})
        df['Similarity'] = similarities
        output = df.sort_values(['Similarity','Prob'],ascending=False).head(10)
        return(output)
autocorrect('gy')
        
```
